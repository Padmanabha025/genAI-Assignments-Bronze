```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model


# Assuming 6 input features and a binary output (fraud or non-fraud)

model = Sequential()

# Input layer (6 features) and first hidden layer
model.add(Dense(units=16, activation='relu', input_dim=6))  # 16 neurons in the hidden layer


model.add(Dense(units=8, activation='relu'))


model.add(Dense(units=1, activation='sigmoid'))  # Sigmoid activation for binary classification


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


plot_model(model, to_file='neural_network_architecture.png', show_shapes=True, show_layer_names=True)


print("Neural Network architecture has been visualized and saved as 'neural_network_architecture.png'.")
```
